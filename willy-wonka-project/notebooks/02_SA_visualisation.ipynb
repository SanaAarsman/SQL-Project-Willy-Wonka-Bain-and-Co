```python
import pandas as pd
import plotly.express as px
df_2 = pd.read_csv("../data/01_location_sales_profit_2.csv")
```

```python
df_2
```

```python

df_2["label"] = df_2["countryregion"] + "<br>" + df_2["region"]

fig = px.scatter(
    df_2,
    x="sales_rank",
    y="profit_percentage_rank",
    text="label",
    size="total_sales",                                  
    title="Sales Rank vs Profit Rank by Region",
    width=900,
    height=600
)

fig.update_traces(textposition="top center")
fig.update_xaxes(autorange="reversed", title="Sales Rank (1 = Highest Sales)")
fig.update_yaxes(autorange="reversed", title="Profit Rank (1 = Highest Profit)")

fig.show()
```
